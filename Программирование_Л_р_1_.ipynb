```python
# Задание task_01_01.
""" Составьте программу, которая запрашивает у пользователя 2 целых числа и выполняет операции:
•	арифметические: +, -, * , / , // , %, **;
•	сравнение: <, <=, >, >=, !=, ==,
выводя на экран результат каждого действия. В случае получение вещественного результата, округлите его до 2-х знаков после запятой (используя функцию round()).
"""
# Выполнил: Князева Анна Павловна
# Группа: ЦИБ-251


a = int(input("a="))
b = int(input("b="))

# Арифметические опреации

print(a + b)
print(a - b)
print(a * b)
print(round(a / b, 2))
print(a // b)
print(a % b)
print(a ** b)

# Опреации сравнения

print(a < b)
print(a <= b)
print(a > b)
print(a >= b)
print(a != b)
print(a == b)
```
